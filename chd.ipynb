```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample data
data = {
    'Sex': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Residence': ['City A', 'City B', 'City A', 'City B', 'City A', 'City B']
}

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Group the data by 'Residence' and 'Sex', and calculate the percentage
grouped_data = df.groupby(['Residence', 'Sex']).size().unstack()
total_per_residence = grouped_data.sum(axis=1)
grouped_data['Male'] = (grouped_data['Male'] / total_per_residence) * 100
grouped_data['Female'] = (grouped_data['Female'] / total_per_residence) * 100


# Get the unique places of residence
residences = grouped_data.index

# Set the width of each bar
bar_width = 0.35

# Create an array of indices for the x-axis ticks
ind = np.arange(len(residences))

# Plotting the clustered chart
fig, ax = plt.subplots()
male_bars = ax.bar(ind, grouped_data['Male'], width=bar_width, label='Male')
female_bars = ax.bar(ind + bar_width, grouped_data['Female'], width=bar_width, label='Female')

# Add labels and title
ax.set_xlabel('Place of Residence')
ax.set_ylabel('Percentage')
ax.set_title('Percentage of Males and Females by Place of Residence')

# Set the x-axis ticks and labels
ax.set_xticks(ind + bar_width / 2)
ax.set_xticklabels(residences)

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

```python
total_per_residence = grouped_data.sum(axis=1)
grouped_data = grouped_data.div(total_per_residence, axis=0) * 100

# Get the unique places of residence
residences = grouped_data.index

# Set the width of each bar
bar_width = 0.35

# Create an array of indices for the x-axis ticks
ind = np.arange(len(residences))

# Plotting the clustered chart
fig, ax = plt.subplots()
male_bars = ax.bar(ind, grouped_data['Male'], width=bar_width, label='Male')
female_bars = ax.bar(ind + bar_width, grouped_data['Female'], width=bar_width, label='Female')

# Add labels and title
ax.set_xlabel('Place of Residence')
ax.set_ylabel('Number of People')
ax.set_title('Sex Division by Place of Residence')

# Set the x-axis ticks and labels
ax.set_xticks(ind + bar_width / 2)
ax.set_xticklabels(residences)

# Add a legend
ax.legend()

# Show the plot
plt.show()
```
